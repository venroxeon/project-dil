```python
# predict_other_columns.py
import pandas as pd
import numpy as np
from lightgbm import LGBMRegressor
import joblib   # optional, for saving models

# ---------------------
# CONFIG
# ---------------------
feature_cols = [
    'Fat', 'Saturated Fats', 'Carbohydrates', 'Sugars', 'Protein',
    'Dietary Fiber', 'Sodium', 'Water', 'Vitamin A', 'Vitamin B11',
    'Vitamin B12', 'Vitamin B6', 'Vitamin C', 'Vitamin D', 'Vitamin E',
    'Vitamin K', 'Calcium', 'Copper', 'Iron', 'Magnesium', 'Phosphorus',
    'Potassium', 'Zinc', 'Nutrition Density'
]

# Minimum training rows required per model
MIN_ROWS = 30

# ---------------------
# 1) Prepare numeric DataFrame with complete rows for training
# ---------------------
df_numeric = df[feature_cols].apply(pd.to_numeric, errors="coerce")
df_complete = df_numeric.dropna()   # only use fully-observed rows for training
print("Rows available for training (complete rows):", df_complete.shape[0])

if df_complete.shape[0] < MIN_ROWS:
    raise RuntimeError(f"Not enough complete rows to train models. Need >= {MIN_ROWS}, got {df_complete.shape[0]}")

# ---------------------
# 2) Train one model per column
# ---------------------
models = {}
for target in feature_cols:
    X_train = df_complete.drop(columns=[target])
    y_train = df_complete[target]

    m = LGBMRegressor(
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=64,
        random_state=42,
        verbose=-1
    )
    m.fit(X_train, y_train)
    models[target] = m

print("Trained models for columns:", list(models.keys()))

# Optional: save models to disk
# for col, mdl in models.items():
#     joblib.dump(mdl, f"model_{col.replace(' ', '_')}.joblib")

# ---------------------
# 3) Iterative prediction function
# ---------------------
def predict_missing_from_one(column_name, value, max_iter=30, tol=1e-3, verbose=False):
    """
    Given column_name and a value, iteratively predict other columns.
    Returns a pandas Series with all feature cols (including the input column set to 'value').
    """
    if column_name not in feature_cols:
        raise ValueError(f"{column_name} not in feature list.")

    # initialize temp vector with column means
    temp = df_complete.mean().to_dict()
    # overwrite with provided value for the known column
    temp[column_name] = float(value)

    temp_prev = temp.copy()

    for it in range(max_iter):
        max_change = 0.0

        # Predict each column (skip the one we already know)
        for targ in feature_cols:
            if targ == column_name:
                continue

            # build input vector for this model: all other features except targ
            input_dict = {c: temp[c] for c in feature_cols if c != targ}
            # Convert to DataFrame for the model
            input_df = pd.DataFrame([input_dict])

            # predict using trained model for targ
            pred = models[targ].predict(input_df)[0]
            # update value; measure absolute change
            change = abs(pred - temp[targ])
            temp[targ] = float(pred)
            if change > max_change:
                max_change = change

        if verbose:
            print(f"iter {it+1}: max change = {max_change:.6f}")

        # check convergence
        if max_change < tol:
            if verbose:
                print("Converged.")
            break

    return pd.Series({c: temp[c] for c in feature_cols})

# ---------------------
# 4) Example usage
# ---------------------
if __name__ == "__main__":
    # Example: user provides Protein=10 (units as in your dataset)
    user_col = "Protein"
    user_val = 10.0

    predicted_row = predict_missing_from_one(user_col, user_val, max_iter=50, tol=1e-4, verbose=True)
    predicted_row[user_col] = user_val  # ensure exact input retained
    print("\nPredicted full row given", f"{user_col}={user_val}")
    print(predicted_row)

    # You can also get a DataFrame with predictions:
    df_pred = predicted_row.to_frame().T
    print("\nAs DataFrame:")
    print(df_pred)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity

def find_similar_rows(predicted_series, df_rows, feature_cols,
                      n_neighbors=5, scale=True, metric="euclidean"):
    """
    predicted_series: pd.Series or dict with keys = feature_cols
    df_rows: DataFrame with rows to search (must contain feature_cols)
    feature_cols: list of feature column names (order matters)
    n_neighbors: how many nearest rows to return
    scale: if True, standardize columns (mean=0, std=1) before measuring distance
    metric: "euclidean" or "cosine"
    """
    # ensure DataFrame and columns
    X = df_rows[feature_cols].astype(float).copy()
    query = pd.Series(predicted_series)[feature_cols].astype(float)

    # scaling (fit scaler on dataset rows only)
    if scale:
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        q_scaled = scaler.transform(query.values.reshape(1, -1))
    else:
        X_scaled = X.values
        q_scaled = query.values.reshape(1, -1)

    if metric == "euclidean":
        dists = euclidean_distances(X_scaled, q_scaled).ravel()
        idx_sorted = np.argsort(dists)
        results = df_rows.iloc[idx_sorted[:n_neighbors]].copy()
        results = results.assign(distance=dists[idx_sorted[:n_neighbors]])
    elif metric == "cosine":
        sims = cosine_similarity(X_scaled, q_scaled).ravel()
        idx_sorted = np.argsort(-sims)  # descending similarity
        results = df_rows.iloc[idx_sorted[:n_neighbors]].copy()
        results = results.assign(cosine_similarity=sims[idx_sorted[:n_neighbors]])
    else:
        raise ValueError("metric must be 'euclidean' or 'cosine'")

    # include original index as a column for traceability
    results = results.reset_index().rename(columns={'index': 'orig_index'})
    return results

# -------------------------
# Example usage with your predicted Series
# -------------------------
# Suppose predicted_series is the pd.Series you printed (Predicted full row)
# If it's in variable `predicted_row` from previous code, use it directly.
# Example: predicted_row = predict_missing_from_one("Protein", 10.0)

# Then run:
nearest = find_similar_rows(predicted_row, df_complete, feature_cols,
                            n_neighbors=5, scale=True, metric="euclidean")

print("Top 5 nearest rows (euclidean, scaled):")
print(nearest)
```
